```
import numpy as np
import mglearn
from sklearn.neighbors import KNeighborsRegressor
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split 
```

```
X,Y = mglearn.datasets.make_wave(n_samples = 50)

#print('Valores de X: ', X)

plt.plot(X,Y,'o')
plt.ylim(Y.min()-1,Y.max()+1)
plt.xlim(X.min()-1,X.max()+1)
plt.ylabel("Target")
plt.xlabel("Input")
```

```
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size = 0.30,random_state=0)

k_range = range(1, 16)
scores = []
scores_t=[]

#se recorre el hyperparameter para validar cual demuestra un mejor resultado
for k in k_range:
    model_knn = KNeighborsRegressor(n_neighbors = k)
    model_knn.fit(X_train, Y_train)
    scores.append(model_knn.score(X_test, Y_test))
    scores_t.append(model_knn.score(X_train, Y_train))

plt.xlabel('n_neighbors')
plt.ylabel('Accuracy')

plt.plot(k_range, scores,label='X_test')
plt.plot(k_range, scores_t,label='X_train')
plt.legend()
```
